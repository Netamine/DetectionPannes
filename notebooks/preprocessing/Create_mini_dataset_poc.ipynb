```python
import pandas as pd
import numpy as np


# Charger les données
df = pd.read_csv(r"..\..\..\Datasources\MetroPT3_imputed_final.csv", delimiter=",", decimal=".", index_col=0)
df.reset_index(drop=True, inplace=True)

# Convertir la colonne timestamp
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
```

```python
display(df.head(3))
```

```python
# Pannes du dataset initial
pannes = [
    {'start': '2020-04-18 00:00:00', 'end': '2020-04-18 23:59:00'},
    {'start': '2020-05-29 23:30:00', 'end': '2020-05-30 06:00:00'},
    {'start': '2020-06-05 10:00:00', 'end': '2020-06-07 14:30:00'},
    {'start': '2020-07-15 14:30:00', 'end': '2020-07-15 19:00:00'},
         ]

# Définir les périodes a garder dans le mini dataset pour la POC
keeped_periods = [
                 {'start': '2020-04-17 23:30:00', 'end': '2020-04-18 00:15:00'},
                 {'start': '2020-05-29 23:00:00', 'end': '2020-05-29 23:45:00'},
                 {'start': '2020-06-05 09:30:00', 'end': '2020-06-05 10:15:00'},
                 {'start': '2020-07-15 14:00:00', 'end': '2020-07-15 14:45:00'},
                 ]


# Construction des données
mini_df = pd.concat([
                    df[(df['timestamp'] >= pd.Timestamp(period['start'])) & 
                   (df['timestamp'] <= pd.Timestamp(period['end']))]
                   for period in keeped_periods
                    ])

print(f"mini_df : {len(mini_df)} lignes")
```

```python
# Distribution des modalités de la variable 'panne' dans l'ensemble d'entrainement
print("Modalités de 'panne' dans l'ensemble du dataframe mini_df :")
print(mini_df["panne"].value_counts())
```

```python
# Convertir la colonne timestamp
mini_df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
```

```python
# Recréer les Timestamp du nouveau dataframe mini_df avec un GAP de 10 S

# Récupérer le premier timestamp de mini_df
start_timestamp = mini_df['timestamp'].iloc[0]

# Générer une nouvelle série de timestamps avec un intervalle de 10 secondes
mini_df['timestamp'] = pd.date_range(start=start_timestamp, periods=len(mini_df), freq='10S')

# Affichage de quelques lignes pour vérification
display(mini_df.head(3))
display(mini_df.tail(3))
```

```python
# Vérification des GAPS

# Calculer la différence entre chaque timestamp consécutif
mini_df['timestamp'] = pd.to_datetime(mini_df['timestamp'])
time_diffs = mini_df['timestamp'].diff().dt.total_seconds().dropna()

# Vérifier si tous les intervalles sont égaux à 10 secondes
if (time_diffs == 10).all():
    print("Tous les intervalles sont de 10 secondes.")
else:
    print("Des écarts irréguliers ont été détectés.")
    print(time_diffs.value_counts())

# Affichage des premieres différences pour validation
print(time_diffs.head(10))
```

```python
# Reconstruction de l'index

# Trier le DataFrame par timestamp croissant
mini_df = mini_df.sort_values(by='timestamp').reset_index(drop=True)

# Récupérer le premier timestamp apres tri
start_timestamp = mini_df['timestamp'].iloc[0]

# Générer une nouvelle série de timestamps avec un intervalle de 10 secondes
mini_df['timestamp'] = pd.date_range(start=start_timestamp, periods=len(mini_df), freq='10S')

# Vérification des premiers timestamps
display(mini_df.head(4))
print("L'index de mini_df a été réinitialisé avec des timestamps réguliers de 10 secondes.")
```

```python
# Enregistrer le dataset miniature sans la colonne panne
mini_df.to_csv(r"..\..\..\Datasources\MetroPT3_miniature.csv", index=True)
```
