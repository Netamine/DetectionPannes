# Stratégie d'imputation : Choix de la méthode d'imputation

<h3>Imputation des Valeurs Manquantes :</h3>
<ul style="font-family: times, serif; font-size:12pt; color:blue;">
    <li><strong>Méthodes Simples :</strong>
        <ul style="color:black;">
            <li>Interpolation linéaire : Estime les valeurs manquantes en calculant une ligne droite entre les points adjacents.</li>
            <li>Valeur moyenne ou médiane : Remplace les données manquantes par la moyenne ou la médiane des observations disponibles.</li>
            <li>Propagation des valeurs : Utilise la derniere valeur observée (forward fill) ou la prochaine valeur disponible (backward fill).</li>
        </ul>
    </li>
    <li><strong>Méthodes Avancées :</strong>
        <ul style="color:black;">
            <li>Modeles de séries temporelles (ARIMA, LSTM) : Exploite les relations temporelles pour prédire les valeurs manquantes.</li>
            <li>K-Nearest Neighbors (KNN) : Trouve des observations similaires et les utilise pour estimer les valeurs manquantes.</li>
            <li>Imputation par régression supervisée : Utilise les relations entre les colonnes pour former un modele prédictif (par ex., régression linéaire, forêt aléatoire).</li>
        </ul>
    </li>
    <li><strong>Approches Basées sur les Données :</strong>
        <ul style="color:black;">
            <li>Saisonnalité et tendances : Détecte des schémas récurrents dans les données pour combler les lacunes.</li>
            <li>Interpolation spline : Utilise des courbes plus complexes pour une meilleure estimation dans des données non linéaires.</li>
            <li>Méthodes spécifiques à un domaine : Exploite des connaissances spécifiques sur le systeme ou le processus pour ajuster les données.</li>
        </ul>
    </li>
    <li><strong>Alternatives Innovantes :</strong>
        <ul style="color:black;">
            <li>Modeles robustes aux données manquantes : Comme les réseaux neuronaux capables de gérer directement les valeurs manquantes.</li>
            <li>Transformation des données manquantes en features : Ajout d'une colonne indiquant les lacunes comme information explicative.</li>
        </ul>
    </li>
</ul>

```python
# Faire une copie du DataFrame
df_imputed_1 = df.copy()

# Appliquer l'interpolation linéaire pour les colonnes continues
for col in cols_continuous:
    df_imputed_1[col] = df_imputed_1[col].interpolate(method='linear')

# Appliquer la propagation des valeurs pour les colonnes catégoriques
for col in cols_categorical:
    df_imputed_1[col] = df_imputed_1[col].fillna(method='ffill')  # Forward fill
    df_imputed_1[col] = df_imputed_1[col].fillna(method='bfill')  # Backward fill
```

```python
# Enregistrer le dataframe corrigé dans un fichier CSV
df_imputed_1.to_csv(r"..\..\..\Datasources/MetroPT3_imputed_1.csv", index=True)
```

```python
# Fonction pour filtrer les lignes contenant des valeurs imputées
def filter_imputed_rows(df):
    # Créer un masque pour détecter les lignes avec des imputations
    mask = df[cols_is_missing].any(axis=1)
    # Filtrer le DataFrame pour afficher toutes les colonnes uniquement pour les lignes imputées
    return df[mask]

# Filtrer et afficher pour chaque méthode
Tmp_df_Only_Imputed_data = filter_imputed_rows(df_imputed_1)

print("Données imputées (Méthode combinée) :")
display(Tmp_df_Only_Imputed_data)

Tmp_df_Only_Imputed_data.describe()
```

<ul style="font-family: times, serif; font-size:14pt; color:green;">
    <li><strong>2. Validation par modele prédictif</strong>
        <ul style="color:black;">
            <li>Notre objectif principal est d'utiliser les données imputées pour former un modele de prédiction de la colonne indiquant les pannes. Pour ce faire, nous comparerons les performances des modeles entrainés à partir des différentes versions du jeu de données apres imputation. Ainsi, chaque modele de régression utilisé pour l'imputation servira de base pour la construction d'un modele de classification. Cette approche nous permettra d'évaluer l'impact des différentes méthodes d'imputation sur la capacité du modele final à prédire efficacement les pannes.</li>
        </ul>
    </li>
</ul>

<ul style="font-family: times, serif; font-size:14pt; color:green;">
    <li><strong>3.  Validation par reconstruction</strong>
        <ul style="color:black;">
            <li>1.Masquer artificiellement des valeurs dans cette portion de données.</li>
            <li>2.Appliquer les méthodes d'imputation.</li>
            <li>3.Comparer les valeurs imputées aux valeurs dorigine.</li>
        </ul>
    </li>
</ul>

```python
# Initialiser des dictionnaires pour stocker les erreurs
errors_combine = {}
errors_mean = {}
errors_median = {}

# Boucle sur chaque colonne
for col in cols_continuous:
    # Masquer 10% des données comme test
    df_test = df.copy()
    mask = np.random.rand(len(df_test)) < 0.1  # Masque pour 10% des données
    df_test.loc[mask, col] = np.nan

    # Appliquer les trois méthodes
    df_test_combine = df_test.copy()
    df_test_combine[col] = df_test_combine[col].interpolate(method='linear')

    df_test_mean = df_test.copy()
    df_test_mean[col] = df_test_mean[col].fillna(df_test_mean[col].mean())

    df_test_median = df_test.copy()
    df_test_median[col] = df_test_median[col].fillna(df_test_median[col].median())

    # Comparer les valeurs imputées avec les valeurs originales
    original_values = df[col][mask]
    combine_values = df_test_combine[col][mask]
    mean_values = df_test_mean[col][mask]
    median_values = df_test_median[col][mask]

    # Calculer les erreurs moyennes pour chaque méthode
    errors_combine[col] = np.mean(np.abs(original_values - combine_values))
    errors_mean[col] = np.mean(np.abs(original_values - mean_values))
    errors_median[col] = np.mean(np.abs(original_values - median_values))


# Créer un DataFrame pour afficher les erreurs
errors_df = pd.DataFrame({
    'Méthode combinée': errors_combine,
    'Imputation Moyenne': errors_mean,
    'Imputation Médiane': errors_median
}).reset_index()

# Renommer les colonnes pour plus de clarté
errors_df.rename(columns={'index': 'Colonnes'}, inplace=True)

# Afficher le DataFrame
display(errors_df)
```

```python
# Colonnes catégoriques à valider
cols_categorical = ['COMP', 'DV_eletric', 'Towers', 'MPG', 'LPS', 'Pressure_switch', 'Oil_level', 'Caudal_impulses']

# Initialiser un dictionnaire pour stocker les précisions
accuracy_results = {'Méthode combinée': {}, 'Imputation Moyenne': {}, 'Imputation Médiane': {}}

# Validation par reconstruction
for col in cols_categorical:
    # Masquer 10% des données dans le DataFrame initial
    mask = np.random.rand(len(df)) < 0.1  # Masquer 10% des données
    original_values = df[col][mask]  # Valeurs originales avant imputation

    # Récupérer les valeurs imputées pour chaque méthode
    combine_values = df_imputed_1[col][mask]
    mean_values = df_imputed_2[col][mask]
    median_values = df_imputed_3[col][mask]

    # Filtrer les NaN (au cas où)
    valid_indices = ~original_values.isna()
    original_values = original_values[valid_indices]
    combine_values = combine_values[valid_indices]
    mean_values = mean_values[valid_indices]
    median_values = median_values[valid_indices]

    # Calculer les précisions pour chaque méthode
    accuracy_results['Méthode combinée'][col] = accuracy_score(original_values, combine_values)
    accuracy_results['Imputation Moyenne'][col] = accuracy_score(original_values, mean_values)
    accuracy_results['Imputation Médiane'][col] = accuracy_score(original_values, median_values)

# Créer un DataFrame pour afficher les précisions
categorical_validation_df = pd.DataFrame(accuracy_results).reset_index()

# Renommer les colonnes
categorical_validation_df.rename(columns={'index': 'Colonnes'}, inplace=True)

# Afficher le DataFrame
display(categorical_validation_df)
```

```python
LSTM (Long Short-Term Memory) est une variante de réseau de neurones récurrents (RNN). Il est conçu pour traiter 
et modéliser des données séquentielles ou temporelles en capturant les dépendances dans les séquences. 
LSTM surmonte les limitations des RNN standards, qui ont du mal à gérer des dépendances à long terme.

Le modele  LSTM fonctionne uniquement pour les colonnes continues:
Les LSTM sont conçus pour modéliser des séries temporelles et capturer les dépendances séquentielles 
entre des valeurs numériques continues.
```

```python
###################################################################################################################### 
### L'imputation par la moyenne est nécessaire avant d'utiliser le LSTM pour garantir des séquences                ###
### continues et éviter les interruptions dues aux valeurs manquantes. Elle assure que le modèle                   ###
### ne rencontre pas de `NaN`,ce qui pourrait entraîner des erreurs de calcul.                                     ###
### Cette approche permet au LSTM d'apprendre efficacement les tendances sans biais causé par les données manquantes.#
######################################################################################################################

import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, accuracy_score
from tensorflow.keras.models import Sequential, save_model, load_model
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras import Input
from tensorflow.keras.callbacks import EarlyStopping
import os
import random

# Démarrer le chronomètre
start_time = time.time()

# Préparer les Données pour LSTM
def prepare_data(series, n_steps):
    """Préparer les séquences pour LSTM."""
    X, y = [], []
    for i in range(len(series) - n_steps):
        seq_x = series[i:i + n_steps]
        seq_y = series[i + n_steps]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)

# Masquer 10 % des données observées uniquement
def mask_random_10_percent(series, original_series):
    """Masquer aléatoirement 10 % des valeurs observées (non manquantes d'origine)."""
    non_missing_indices = original_series.dropna().index
    mask_indices = random.sample(list(non_missing_indices), int(0.1 * len(non_missing_indices)))
    series_masked = series.copy()
    series_masked.loc[mask_indices] = np.nan
    return series_masked, mask_indices

# Créer un répertoire pour sauvegarder les modeles
model_dir = "lstm_models"
os.makedirs(model_dir, exist_ok=True)

# Initialisation
df_imputed_10 = df.copy()  # Créer un DataFrame pour l'imputation
scaler = MinMaxScaler(feature_range=(0, 1))
n_steps = 10  # Fenêtre temporelle
results_continuous = {}
results_categorical = {}

# Colonnes continues
cols_continuous = ['TP2', 'TP3', 'H1', 'DV_pressure', 'Reservoirs', 'Oil_temperature', 'Motor_current']

# Colonnes catégoriques ou binaires
cols_categorical = ['COMP', 'DV_eletric', 'Towers', 'MPG', 'LPS', 'Pressure_switch', 'Oil_level', 'Caudal_impulses']

# Callback pour Early Stopping
early_stopping = EarlyStopping(
    monitor='loss',  # Surveiller la fonction de perte
    patience=3,      # Arrêter apres 3 epochs sans amélioration
    restore_best_weights=True  # Restaurer les meilleurs poids
)

# **** Traitement des Colonnes Continues ****
print("**** Traitement des Colonnes Continues ****")

for col in cols_continuous:
    print(f"Imputation et Validation pour la colonne : {col}")
    
    # Remplacement des NaN par la Moyenne
    series = df_imputed_10[col].copy()
    original_series = df[col].copy()
    print(f"Nombre de NaN dans la colonne {col} avant remplacement : {series.isna().sum()}")
    series.fillna(series.mean(), inplace=True)

    # Masquage pour Validation
    series_masked, mask_indices = mask_random_10_percent(series, original_series)

    # Normalisation
    series_scaled = scaler.fit_transform(series_masked.values.reshape(-1, 1))

    # Préparer les Données pour LSTM
    X, y = prepare_data(series_scaled.flatten(), n_steps)
    X = X.reshape((X.shape[0], X.shape[1], 1))
    valid_indices = ~np.isnan(X).any(axis=(1, 2)) & ~np.isnan(y)
    X = X[valid_indices]
    y = y[valid_indices]
    print(f"Nombre de séquences valides pour {col} apres filtrage : {len(X)}")

    if len(X) == 0:
        print(f"Aucune séquence valide pour l'entrainement pour la colonne {col}.")
        continue

    # Construire et Entrainer le Modele LSTM
    model = Sequential()
    model.add(Input(shape=(n_steps, 1)))
    model.add(LSTM(50, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X, y, epochs=20, verbose=1, callbacks=[early_stopping])

    # Sauvegarder le Modele au format .keras
    model_path = os.path.join(model_dir, f"model_{col}.keras")
    model.save(model_path)
    print(f"Modele pour la colonne {col} sauvegardé à : {model_path}")

    # Validation
    valid_indices = [idx for idx in mask_indices if idx >= n_steps]
    if valid_indices:
        sequences_validation = [
            series_scaled[idx - n_steps:idx].reshape((1, n_steps, 1))
            for idx in valid_indices
        ]
        if sequences_validation:
            sequences_validation = np.vstack(sequences_validation)
            predictions_validation = model.predict(sequences_validation, verbose=0)
            predicted_values = scaler.inverse_transform(predictions_validation).flatten()
            true_values = original_series.loc[valid_indices]

            # Filtrer les NaN et aligner les indices
            valid_mask = ~np.isnan(predicted_values) & ~np.isnan(true_values)
            predicted_values = predicted_values[valid_mask]
            true_values = true_values[valid_mask]

            if len(true_values) > 0 and len(predicted_values) > 0:
                # Calculer les métriques
                rmse = np.sqrt(mean_squared_error(true_values, predicted_values))
                mae = mean_absolute_error(true_values, predicted_values)
                results_continuous[col] = {'RMSE': rmse, 'MAE': mae}
                print(f"Validation terminée pour {col} : RMSE = {rmse:.4f}, MAE = {mae:.4f}")
            else:
                print(f"Aucune valeur valide pour calculer les métriques pour {col}.")
        else:
            print(f"Aucune séquence valide pour la validation pour {col}.")
    else:
        print(f"Pas assez de données pour valider la colonne {col}.")

# **** Traitement des Colonnes Catégoriques ****
print("\n**** Traitement des Colonnes Catégoriques ****")

for col in cols_categorical:
    print(f"Imputation pour la colonne catégorique : {col}")
    # Remplacement des NaN par la Modalité la Plus Fréquente (Mode)
    series = df_imputed_10[col].copy()
    mode_value = series.mode().iloc[0]
    series.fillna(mode_value, inplace=True)
    df_imputed_10[col] = series
    print(f"Valeur imputée pour {col} : {mode_value}")

# Résultats
print("\nRésultats des Colonnes Continues :")
for col, metrics in results_continuous.items():
    print(f"{col} : RMSE = {metrics['RMSE']:.4f}, MAE = {metrics['MAE']:.4f}")

print("\nRésultats des Colonnes Catégoriques :")
for col, metrics in results_categorical.items():
    print(f"{col} : Accuracy = {metrics['Accuracy']:.4f}")

# Arrêter le chronomètre
end_time = time.time()
print(f"Temps d'exécution: {end_time - start_time:.2f} secondes")
```

```python
# Prédiction des données manquantes
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import load_model
import os


df_sample = pd.read_csv(r"..\Generated_Files\captures_30mn.csv")

# Configuration
model_dir = r"..\Generated_Files\lstm_models"  # Répertoire contenant les modeles sauvegardés 
n_steps = 10  # Taille de la fenêtre temporelle
scaler = MinMaxScaler(feature_range=(0, 1))

# Imputation avec les modeles sauvegardés
for col in cols_continuous:
    print(f"Imputation pour la colonne : {col}")
    
    # Charger le modele LSTM correspondant
    model_path = os.path.join(model_dir, f"model_{col}.keras")
    if not os.path.exists(model_path):
        print(f"Modele introuvable pour la colonne {col}.")
        continue
    
    model = load_model(model_path)
    print(f"Modele chargé pour la colonne {col} : {model_path}")
    
    # Normaliser la colonne
    series = df_sample[col].copy()
    series.fillna(series.mean(), inplace=True)  # Remplacer temporairement les NaN
    series_scaled = scaler.fit_transform(series.values.reshape(-1, 1))
    
    # Identifier les indices manquants
    missing_indices = [idx for idx in df_sample[col].isna().index if idx >= n_steps]
    print(f"Nombre de valeurs manquantes valides pour {col} : {len(missing_indices)}")
    
    # Préparer les séquences pour les valeurs manquantes
    sequences = [
        series_scaled[idx - n_steps:idx].reshape((1, n_steps, 1))
        for idx in missing_indices
    ]
    
    if sequences:
        sequences = np.vstack(sequences)

        # Prédire les valeurs manquantes
        predictions = model.predict(sequences, batch_size=512, verbose=1)
        predicted_values = scaler.inverse_transform(predictions).flatten()

        # Remplir les valeurs manquantes
        for i, idx in enumerate(missing_indices):
            df_sample.loc[idx, col] = predicted_values[i]

        print(f"Imputation terminée pour {col}.")
    else:
        print(f"Pas assez de données pour imputer les valeurs manquantes pour {col}.")
```
